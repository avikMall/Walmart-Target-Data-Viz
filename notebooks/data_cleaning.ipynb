```python
import pandas as pd

target = pd.read_csv('target_final.csv')
walmart = pd.read_csv('walmart_final.csv')
census = pd.read_csv('county.csv')
```

```python
walmart.shape
```

```python
for index, value in enumerate(census['County'].values):
    if value[-6:] == 'County':
        value = value[:-7]
        census['County'].values[index] = value
census.head()
```

```python
census['County'] = census['County'].str.strip().str.lower()
target['Address.County'] = target['Address.County'].str.strip().str.lower()
census.shape
```

```python
census_deduped = census.drop_duplicates(subset=['County'], keep='first')
census_deduped.shape
```

```python
merged_data = pd.merge(target, census_deduped, left_on=['Address.County'], right_on=['County'], how='left')
merged_data.shape
```

```python
merged_data.to_csv('target_final.csv', index=False)
```

```python
merged_data_2 = pd.merge(walmart, census_deduped, left_on=['COUNTYNAME'], right_on=['County'], how='left')
merged_data_2.shape
```

```python
merged_data_2.head()
```

```python
merged_data_2.to_csv('walmart_final.csv')
```
